```python
import pprint
from collections import defaultdict
from gensim.parsing.preprocessing import STOPWORDS
from gensim import corpora, models, similarities
```

```python
def normalize(corpus):
    return [[word for word in document.lower().split() if word not in STOPWORDS] for document in corpus]
```

```python
def freq_threshold(corpus, threshold):
    # Count word frequencies
    frequency = defaultdict(int)
    for text in corpus:
        for token in text:
            frequency[token] += 1
    # Only keep words appearing more than threshold
    processed_corpus = [[token for token in text if frequency[token] > threshold] for text in corpus]
    return processed_corpus
```

```python
# Normalize pr's and did's - lower, split and remove stopwords
prs_norm = normalize(prs)
dids_norm = normalize(dids)
```

```python
# Only keep words appearing more than defined threshold
processed_corpus = freq_threshold(prs_norm, 0)
pprint.pprint(processed_corpus)
```

```python
# Associate each word in corpus with a unique int ID
dictionary = corpora.Dictionary(processed_corpus)
pprint.pprint(dictionary.token2id)
```

```python
# Create bag of words representation of corpus
bow_corpus = [dictionary.doc2bow(text) for text in processed_corpus]
pprint.pprint(bow_corpus)
```

```python
# Train tf-idf model
tfidf = models.TfidfModel(bow_corpus)
# Transform whole corpus via tf-idf and index it
index = similarities.SparseMatrixSimilarity(tfidf[bow_corpus], num_features=93, num_best=1)
# Query the similarity of "query document" against every document in corpus
sims = []
count = 0
for query_document, i in zip(dids_norm, range(len(dids))):
    query_bow = dictionary.doc2bow(query_document)
    sim = index[tfidf[query_bow]]
    if sim != []:
        # Only print similarity score greater than x%, does NOT imply correctness
        if sim[0][1] > 0.20:
            print('Associated: ', dids[i], '\nwith PR: ', prs[sim[0][0]], '\nSimilarity %: ', sim[0][1], '\n\n')
            count += 1

print('Commits which met threshold: {}/{}'.format(count, len(dids)), '(%{})'.format(round(count/len(dids), 2)))
```
