```python
import json


with open('raw.json') as f:
    data = json.load(f)

data.keys()
```

```python
from bs4 import BeautifulSoup


soup = BeautifulSoup(data['full'], 'html.parser')
```

```python
print(soup.prettify())
```

```python
import pandas as pd

def item_filter(tag):
    cond = tag.name == 'li'
    cond = cond and 'PR' in tag.text
    cond = cond and tag.find('a').get('href')
    return cond


data = []
for it in soup.find_all(item_filter):
    data.append({
        'text': it.text,
        'link': it.find('a').get('href'),
    })
df = pd.DataFrame(data)
```

```python
df
```

```python
state_map = {
    '‚ö†': 'wip',
    'üëÄ': 'needs review',
    '‚úî': 'done',
}

df['state'] = df.text.apply(lambda x: state_map[x[0]])
df
```

```python
def extract_contributors(text):
    contributors = text.split('(')[-1].split(')')[0]
    if text == contributors:
        return ''
    return contributors

df['contributors'] = df.text.apply(extract_contributors)
df
```

```python
df.iloc[25].text
```

```python
def clean(text):
    _, _, _, text = text.split(' ', maxsplit=3)
    text = text.split('(')[0]
    return text

df['text_cleaned'] = df.text.apply(clean)
df
```
